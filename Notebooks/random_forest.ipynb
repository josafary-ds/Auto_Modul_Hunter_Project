```python
import pickle
from sklearn.metrics import accuracy_score, classification_report
from yellowbrick.classifier import ConfusionMatrix
from sklearn.ensemble import RandomForestClassifier
```

```python
with open('../feets_data.pkl', 'rb') as file:
    x_feets_training, x_feets_test, y_feets_training, y_feets_test = pickle.load(file)
```

```python
 x_feets_training.shape, y_feets_training.shape
```

```python
x_feets_test.shape, y_feets_test.shape
```

## Creating the Forest

```python
rforest_feets = RandomForestClassifier(n_estimators=70, criterion='entropy', random_state=0)
rforest_feets.fit(x_feets_training, y_feets_training)
```

```python
rforest_predict = rforest_feets.predict(x_feets_test)
rforest_predict
```

## Comparing to real data

```python
accuracy_score(y_feets_test, rforest_predict)
```

```python
cm = ConfusionMatrix(rforest_feets)
cm.fit(x_feets_training, y_feets_training)
cm.score(x_feets_test, y_feets_test)
```

```python
print(classification_report(y_feets_test, rforest_predict))
```
