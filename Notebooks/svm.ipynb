```python
import pickle
from sklearn.metrics import accuracy_score, classification_report
from yellowbrick.classifier import ConfusionMatrix
from sklearn.svm import SVC
```

```python
with open('../feets_data.pkl', 'rb') as file:
    x_feets_training, x_feets_test, y_feets_training, y_feets_test = pickle.load(file)
```

```python
svm_feets = SVC(kernel='linear', random_state=1, C=1.0)
svm_feets.fit(x_feets_training, y_feets_training)
```

```python
svm_predict = svm_feets.predict(x_feets_test)
svm_predict
```

```python
accuracy_score(y_feets_test, svm_predict)
```

```python
cm = ConfusionMatrix(svm_feets)
cm.fit(x_feets_training, y_feets_training)
cm.score(x_feets_test, y_feets_test)
```
