```python
from sklearn.neural_network import MLPClassifier
import pickle
from sklearn.metrics import accuracy_score, classification_report
from yellowbrick.classifier import ConfusionMatrix
```

```python
with open('../feets_data.pkl', 'rb') as file:
    x_feets_training, x_feets_test, y_feets_training, y_feets_test = pickle.load(file)
```

```python
x_feets_training.shape, x_feets_test.shape
```

```python
y_feets_training.shape, y_feets_test.shape
```

## Creating the neural network

```python
neural_net_feets = MLPClassifier(verbose=True, max_iter=500, tol=0.000010, hidden_layer_sizes=(7, 7))
neural_net_feets.fit(x_feets_training, y_feets_training)
```

```python
neural_net_predict = neural_net_feets.predict(x_feets_test)
neural_net_predict
```

```python
accuracy_score(y_feets_test, neural_net_predict)
```

```python
cm = ConfusionMatrix(neural_net_feets)
cm.fit(x_feets_training, y_feets_training)
cm.score(x_feets_test, y_feets_test)
```

```python
print(classification_report(y_feets_test, neural_net_predict))
```
